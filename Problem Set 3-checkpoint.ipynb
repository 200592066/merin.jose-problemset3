```python
# Question 1
    
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
# Assuming the dataset is in CSV format, adjust accordingly if it's in a different format
users = pd.read_csv(url, sep='|', index_col='user_id')

# Step 4: Discover the mean age per occupation
mean_age_by_occupation = users.groupby('occupation')['age'].mean()

# Step 5: Discover the Male ratio per occupation and sort it
male_ratio_by_occupation = users.groupby('occupation')['gender'].apply(lambda x: (x == 'M').sum() / x.count()).sort_values(ascending=False)

# Step 6: For each occupation, calculate the minimum and maximum ages
min_max_age_by_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])

# Step 7: For each combination of occupation and sex, calculate the mean age
mean_age_by_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean()

# Step 8: For each occupation, present the percentage of women and men
gender_percentage_by_occupation = users.groupby(['occupation', 'gender']).size().unstack().apply(lambda x: x / x.sum(), axis=1)

# Display results
print("Mean Age per Occupation:\n", mean_age_by_occupation, "\n")
print("Male Ratio per Occupation (from most to least):\n", male_ratio_by_occupation, "\n")
print("Minimum and Maximum Ages per Occupation:\n", min_max_age_by_occupation, "\n")
print("Mean Age per Occupation and Sex:\n", mean_age_by_occupation_sex, "\n")
print("Percentage of Women and Men per Occupation:\n", gender_percentage_by_occupation)
```

```python
# Question 4

# Step 1: Import necessary libraries
import pandas as pd

# Step 2: Import the dataset from the attached file wind.txt
url = "wind.txt"
data = pd.read_csv(url, delim_whitespace=True)

# Step 3: Replace the first 3 columns by a proper datetime index
data['date'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1))
data = data.set_index('date').drop(['Yr', 'Mo', 'Dy'], axis=1)

# Step 4: Create a function to fix the year and apply it
def fix_year(x):
    if x.year > 2000:
        year = x.year - 100
    else:
        year = x.year
    return pd.to_datetime(f"{year}-{x.month}-{x.day}")

data.index = data.index.map(fix_year)

# Step 5: Set the right dates as the index with data type datetime64[ns]
data.index = pd.to_datetime(data.index)

# Step 6: Count missing values for each location over the entire record
missing_values_location = data.isnull().sum()

# Step 7: Count non-missing values in total
non_missing_values_total = data.notnull().sum().sum()

# Step 8: Calculate the mean windspeeds over all locations and times
mean_windspeeds = data.mean().mean()

# Step 9: Calculate min, max, mean, and standard deviations at each location
loc_stats = data.describe().transpose()[['min', 'max', 'mean', 'std']]

# Step 10: Calculate min, max, mean, and standard deviations for each day
day_stats = data.resample('D').agg(['min', 'max', 'mean', 'std'])

# Step 11: Find the average windspeed in January for each location
january_avg = data[data.index.month == 1].groupby(data.index.year).mean()

# Step 12: Downsample to a yearly frequency for each location
yearly_freq = data.resample('Y').mean()

# Step 13: Downsample to a monthly frequency for each location
monthly_freq = data.resample('M').mean()

# Step 14: Downsample to a weekly frequency for each location
weekly_freq = data.resample('W').mean()

# Step 15: Calculate min, max, mean, and standard deviations across all locations for each week (first 52 weeks)
weekly_stats = data.resample('W-Mon', closed='right', label='right').agg(['min', 'max', 'mean', 'std']).head(52)

# Display results (you may want to print or explore the results further)
print("Missing values for each location:\n", missing_values_location, "\n")
print("Non-missing values in total:", non_missing_values_total, "\n")
print("Mean windspeeds over all locations and times:", mean_windspeeds, "\n")
print("Location statistics (min, max, mean, std):\n", loc_stats, "\n")
print("Day statistics (min, max, mean, std):\n", day_stats, "\n")
print("Average windspeed in January for each location:\n", january_avg, "\n")
print("Yearly frequency for each location:\n", yearly_freq, "\n")
print("Monthly frequency for each location:\n", monthly_freq, "\n")
print("Weekly frequency for each location (first 52 weeks):\n", weekly_freq.head(52))
```

```python
 # Question 5
    
    # Step 1: Import necessary libraries
import pandas as pd

# Step 2: Import the dataset from the provided address
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

# Step 3: Assign it to a variable called chipo

# Step 4: See the first 10 entries
print("First 10 entries:\n", chipo.head(10), "\n")

# Step 5: What is the number of observations in the dataset?
num_observations = len(chipo)

# Step 6: What is the number of columns in the dataset?
num_columns = chipo.shape[1]

# Step 7: Print the name of all the columns.
column_names = chipo.columns.tolist()

# Step 8: How is the dataset indexed?
index_type = chipo.index

# Step 9: Which was the most-ordered item?
most_ordered_item = chipo['item_name'].value_counts().idxmax()

# Step 10: For the most-ordered item, how many items were ordered?
most_ordered_item_quantity = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()

# Step 11: What was the most ordered item in the choice_description column?
most_ordered_choice_description = chipo['choice_description'].value_counts().idxmax()

# Step 12: How many items were ordered in total?
total_ordered_items = chipo['quantity'].sum()

# Step 13: Turn the item price into a float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

# Check the item price type before and after
item_price_type_before = chipo['item_price'].dtype

# Create a lambda function and change the type of item price
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x))

# Check the item price type after
item_price_type_after = chipo['item_price'].dtype

# Step 14: How much was the revenue for the period in the dataset?
revenue = (chipo['quantity'] * chipo['item_price']).sum()

# Step 15: How many orders were made in the period?
num_orders = chipo['order_id'].nunique()

# Step 16: What is the average revenue amount per order?
average_revenue_per_order = revenue / num_orders

# Step 17: How many different items are sold?
num_different_items = chipo['item_name'].nunique()

# Display results
print("Number of observations in the dataset:", num_observations)
print("Number of columns in the dataset:", num_columns)
print("Column names:", column_names)
print("Dataset indexed as:", index_type)
print("Most-ordered item:", most_ordered_item)
print("Quantity of the most-ordered item:", most_ordered_item_quantity)
print("Most ordered choice description:", most_ordered_choice_description)
print("Total ordered items:", total_ordered_items)
print("Item price type before conversion:", item_price_type_before)
print("Item price type after conversion:", item_price_type_after)
print("Revenue for the period:", revenue)
print("Number of orders:", num_orders)
print("Average revenue amount per order:", average_revenue_per_order)
print("Number of different items sold:", num_different_items)
```

```python
# Question 6

# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Assuming you have the data in a DataFrame named 'data'
# Replace 'your_dataset.csv' with the actual file path or URL
url = "us_marriage.csv"
data = pd.read_csv(url)

# Step 3: Assuming your dataset has columns 'Year', 'Marriages', 'Divorces', and 'Population'
# Adjust the column names accordingly if needed

# Step 4: Calculate marriages and divorces per capita
data['Marriages_Per_Capita'] = data['Marriages'] / data['Population']
data['Divorces_Per_Capita'] = data['Divorces'] / data['Population']

# Step 5: Create a line plot
plt.figure(figsize=(12, 6))

plt.plot(data['Year'], data['Marriages_Per_Capita'], label='Marriages per Capita', color='blue')
plt.plot(data['Year'], data['Divorces_Per_Capita'], label='Divorces per Capita', color='red')

# Step 6: Label the axes and title
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Number of Marriages and Divorces per Capita in the U.S. (1867-2014)')

# Step 7: Show the legend
plt.legend()

# Step 8: Display the plot
plt.show()
```

```python
# Question 8

# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Create a sample dataset (replace this with your actual data)
data = {
    'Actor': ['Arnold Schwarzenegger', 'Sylvester Stallone', 'Keanu Reeves', 'Tom Cruise'],
    'Kill_Count': [200, 180, 150, 120]
}

actors_df = pd.DataFrame(data)

# Step 3: Sort the actors by their kill count
sorted_actors = actors_df.sort_values(by='Kill_Count', ascending=True)

# Step 4: Create a horizontal bar chart
plt.figure(figsize=(10, 6))

plt.barh(sorted_actors['Actor'], sorted_actors['Kill_Count'], color='red')

# Step 5: Label the axes and title
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')

# Step 6: Display the actor names on each bar
for index, value in enumerate(sorted_actors['Kill_Count']):
    plt.text(value, index, str(value))

# Step 7: Display the plot
plt.show()
```

```python
# Question 9

# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Sample data (replace this with your actual data)
data = {
    'Emperor': ['Augustus', 'Caligula', 'Nero', 'Trajan', 'Marcus Aurelius'],
    'Fate': ['Natural Death', 'Assassinated', 'Assassinated', 'Natural Death', 'Natural Death']
}

emperors_data = pd.DataFrame(data)

# Step 3: Calculate the fraction of emperors assassinated
assassination_counts = emperors_data['Fate'].value_counts()
total_emperors = len(emperors_data)

assassination_fraction = assassination_counts.get('Assassinated', 0) / total_emperors

# Step 4: Create a pie chart
labels = ['Assassinated', 'Natural Death']
sizes = [assassination_fraction, 1 - assassination_fraction]
colors = ['red', 'green']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Step 5: Make the pie chart an even circle
plt.axis('equal')

# Step 6: Title the pie chart
plt.title('Fraction of Roman Emperors Assassinated')

# Step 7: Display the pie chart
plt.show()
```
